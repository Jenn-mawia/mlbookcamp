# Multiplication

## Vector-vector multiplication (dot product)



Screenshot 2023-04-24 at 12.53.10.png

```python
def vector_vector_multiplication(u,v):
  assert u.shape[0] == v.shape[0] #check that the vectors have same size(shape)
  n = u.shape[0] #number of elements in the vectors
  result = 0.0 
  for i in range(n): 
    result = result + u[i] * v[i] 
  return result
```

```python
vector_vector_multiplication(u,v)
```

```python
u.dot(v) #same formula from numpy
```

## Matrix-vector multiplication

Screenshot 2023-04-24 at 12.58.10.png

```python
#matrix U
U = np.array([
    [2,4,5,6],
    [1,2,1,2],
    [3,1,2,1],
])
U
```

```python
def matrix_vector_multiplication(U,v):
  assert U.shape[1] == v.shape[0] #ensure dimensionality matches
  num_rows = U.shape[0]

  result = np.zeros(num_rows) #three rows in the matrix U
  for i in range(num_rows):
    result[i] = vector_vector_multiplication(U[i], v)
  return result
```

```python
matrix_vector_multiplication(U,v)
```

```python
# numpy version of matrix vector multiplication
U.dot(v)
```

## Matrix-Matrix multiplication

Screenshot 2023-04-24 at 13.07.21.png

```python
V = np.array([
    [1,1,2],
    [0,0.5,1],
    [0,2,1],
    [2,1,0],
])
V
```

```python
def matrix_matrix_multiplication(U, V):
  assert U.shape[1] == V.shape[0]
  num_rows = U.shape[0]
  num_cols = V.shape[1]
  result = np.zeros((num_rows, num_cols)) 

  for i in range(num_cols):
    vi = V[:,i]
    Uvi = matrix_vector_multiplication(U, vi)
    result[:,i] = Uvi

  return result
```

```python
matrix_matrix_multiplication(U,V)
```

```python
U.dot(V)
```
